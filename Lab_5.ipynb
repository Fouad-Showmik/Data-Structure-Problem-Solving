```python
 # TASK 1
class Stack:
    def __init__(self):
        self.stack = []
        self.pointer = -1

    def push(self, a):
        self.stack.append(a)
        self.pointer += 1

    def pop(self):
        if not self.stack:
            print("Empty")
        else:
            self.stack = self.stack[:-1]
            self.pointer -= 1

    def peek(self):
        if not self.stack:
            return "Empty stack"
        else:
            return self.stack[-1]


def isValid(expression):
    stack1 = Stack()
    stack2 = Stack()
    opening_bracket = ['(', '{', '[']
    closing_bracket = [')', '}', ']']
    res = 0
    check = True

    for i in range(len(expression)):
        if expression[i] in opening_bracket:
            stack1.push(expression[i])
            stack2.push(i)
            res += 1

        elif expression[i] in closing_bracket:
            if res == 0:
                print("This expression is NOT correct.")
                print("Error at character # {}. '{}' not {}.".format(i + 1, expression[i], "opened"))
                check = False
                break
            else:
                top_character = stack1.peek()
                top_position = stack2.peek()
                if top_character == '[' and expression[i] == ']':
                    stack1.pop()
                    stack2.pop()
                    res -= 1
                elif top_character == '{' and expression[i] == '}':
                    stack1.pop()
                    stack2.pop()
                    res -= 1
                elif top_character == '(' and expression[i] == ')':
                    stack1.pop()
                    stack2.pop()
                    res -= 1
                else:
                    print("This expression is NOT correct.")
                    print("Error at character # {}. '{}' not {}.".format(top_position + 1, top_character, "closed"))
                    check = False
                    break

    if res == 0 and check is True:
        print("This expression is correct.")

expression = input("Enter an arithmetic expression:")
isValid(expression)
```

```python
 # TASK 2
class Node:
    def __init__(self, element):
        self.element = element
        self.next = None


class Stack:
    def __init__(self):
        self.head = None

    def push(self, a):
        if self.head is None:
            self.head = Node(a)
        else:
            new_node = Node(a)
            new_node.next = self.head
            self.head = new_node

    def pop(self):
        if self.head is None:
            print("Empty")
        else:
            temp = self.head
            self.head = self.head.next
            temp.element = None
            temp.next = None

    def peek(self):
        return self.head.element


def isValid(expression):
    stack1 = Stack()
    stack2 = Stack()
    opening_bracket = ['(', '{', '[']
    closing_bracket = [')', '}', ']']
    res = 0
    check = True

    for i in range(len(expression)):
        if expression[i] in opening_bracket:
            stack1.push(expression[i])
            stack2.push(i)
            res += 1

        elif expression[i] in closing_bracket:
            if res == 0:
                print("This expression is NOT correct.")
                print("Error at character # {}. '{}' not {}.".format(i + 1, expression[i], "opened"))
                check = False
                break
            else:
                top_character = stack1.peek()
                top_position = stack2.peek()
                if top_character == '[' and expression[i] == ']':
                    stack1.pop()
                    stack2.pop()
                    res -= 1
                elif top_character == '{' and expression[i] == '}':
                    stack1.pop()
                    stack2.pop()
                    res -= 1
                elif top_character == '(' and expression[i] == ')':
                    stack1.pop()
                    stack2.pop()
                    res -= 1
                else:
                    print("This expression is NOT correct.")
                    print("Error at character # {}. '{}' not {}.".format(top_position + 1, top_character, "closed"))
                    check = False
                    break

    if res == 0 and check is True:
        print("This expression is correct.")

expression = input("Enter an arithmetic expression:")
isValid(expression)
```
